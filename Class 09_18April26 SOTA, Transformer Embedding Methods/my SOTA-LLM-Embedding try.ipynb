```python
# 1. Dot Product
# 2. Cosine Similarity
# # 3. Eculidean Distance(L2)
```

```python
# Formulas for Dot Product 

def dot_product(a, b):
    return np.dot(a, b)
```

```python
# Formulas for cosine_similarity


def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
# Formulas for euclidean_distance


def euclidean_distance(a, b):
    a = np.array(a)
    b = np.array(b)
    return np.linalg.norm(a - b)
```
